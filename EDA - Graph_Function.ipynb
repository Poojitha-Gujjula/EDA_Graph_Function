# 1. Graph_V1

Create a function named Graph_v1 that takes a dataset as input and exports boxplots and
histograms (in the form of PNG files) for all the numerical columns in the data and bar plots for all
the categorical columns of the data.

```python
def graph_v1(file):
    dat=pd.read_csv(file)
    for i in dat.columns:
        if i in dat.loc[:,dat.dtypes==object]:
            dat[i].value_counts().plot(kind='bar', figsize=(10,8), color='turquoise', fontsize=12, rot=0)
            plt.xlabel(i,fontsize=15)
            plt.ylabel('Frequency',fontsize=15)
            fig=plt.gcf() #This is helping us save the figure 
            plt.show()
            fig.savefig(i+'.jpg',dpi=100) #Tried getting png images but we are getting distroted images 
            #Hence saving as jpg images 
        else:
            plt.subplot(2,1,2)
            plt.hist(dat[i],histtype='bar',bins=20,color='turquoise',edgecolor='white')
            plt.xlabel(i)
            plt.ylabel('Frequency')

            plt.subplot(2,1,1)
            plt.boxplot(dat[i], vert=False)
            plt.title(f'Distribution of {i}')
            fig1=plt.gcf()
            plt.show()
            fig1.savefig(i+'.jpg',dpi=100)
```

```python
graph_v1(r'C:\Users\pooji\Downloads\attrition.csv')
```

# 2. Graph_V2

Create a function named Graph_v2 that takes a dataset as input and exports bar plots for all the
categorical columns of the data and for the numeric columns it checks if the if the variable is discrete
(or contains a lesser number of unique values) or continuous and plot the graph accordingly. For
discrete numeric variables, it plots bar plot and for continuous it plots histogram &amp; boxplot.

```python
import os
def graph_v2(file,dir=r'C:\Users\pooji\Desktop\python\graph2'):
    dat=pd.read_csv(file)
    import os
    os.chdir(dir)
    for i in dat.columns:
        if i in dat.loc[:,dat.dtypes==object] or dat[i].value_counts().count()<10: 
            #Here we are trying to get a bar plot for categorical as well as discrete numbers <10
            dat[i].value_counts().plot(kind='bar', figsize=(10,8), color='turquoise', fontsize=12, rot=0)
            plt.xlabel(i,fontsize=15)
            plt.ylabel('Frequency',fontsize=15)
            fig=plt.gcf()
            plt.show()
            fig.savefig(i+'.jpg',dpi=100)
        else:
            plt.subplot(2,1,2)
            plt.hist(dat[i],histtype='bar',bins=20,color='turquoise',edgecolor='white')
            plt.xlabel(i)
            plt.ylabel('Frequency')

            plt.subplot(2,1,1)
            plt.boxplot(dat[i], vert=False)
            plt.title(f'Distribution of {i}')
            fig1=plt.gcf()
            plt.show()
            fig1.savefig(i+'.jpg',dpi=100)
```

```python
graph_v2(r'C:\Users\pooji\Downloads\attrition.csv')
```
